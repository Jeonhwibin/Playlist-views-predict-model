```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.family'] = 'AppleGothic'  # 맥북에서 사용 가능한 한글 폰트
plt.rcParams['axes.unicode_minus'] = False  # 마이너스(-) 기호 깨짐 방지
```

```python
df_1 = pd.read_csv('raw_data/youtube_info/구독자_조회수_추가_log_linear(최종)/rreal_final_all_youtubeAPI_info.csv', encoding='utf-8') # 기준 개수
df_2 = pd.read_csv('raw_data/thumbnails/thumbnails_hundred_text.csv', encoding='utf-8')
df_3 = pd.read_csv('raw_data/thumbnails/thumbnails_colors.csv', encoding='utf-8')
df_4 = pd.read_csv('raw_data/thumbnails/thumbnails_face_analysis_YOLOv8.csv', encoding='utf-8') # 정면 얼굴 비율 기준 0.15로 낮춘 것
df_5 = pd.read_csv('raw_data/audio_features/final_audio_features.csv', encoding='utf-8')
df_6 = pd.read_csv('raw_data/title_text/final_title_text.csv', encoding='utf-8')
df_7 = pd.read_csv('raw_data/thumbnails/thumbnails_objects.csv', encoding='utf-8')
```

```python
import pandas as pd

# 예시: 문자열 쉼표로 나눠서 해시태그 수 세기
df_1['hashtag_count'] = df_1['hashtags'].fillna('').apply(
    lambda x: len([tag.strip() for tag in x.split(',') if tag.strip()])
)
df_1['hashtag_count']
```

```python
df_1 = df_1.loc[:, [
    'video_id', 'publish_date',
    'upload_elapsed_0511', 'upload_elapsed_0521', 'upload_elapsed_0522',
    'subscriber_count_0511', 'subscriber_count_0521', 'subscriber_count_0522',
    'log_linear_subs_180d_1', 'log_linear_subs_180d_2',
    'views_0511', 'views_0521', 'views_0522',
    'log_linear_views_180d_1', 'log_linear_views_180d_2',
    'linear_views_180d_1', 'linear_views_180d_2',
    'linear_subs_180d_1', 'linear_subs_180d_2',
    # 'viral_views_180d', 'nonviral_views_180d',
    'hashtag_count'
]]
df_2 = df_2
df_3 = df_3
df_4 = df_4
df_5 = df_5.loc[:, [
    'video_id', 'pitch_mean', 'energy_mean', 'centroid_mean', 'bpm',
    'speech_rate', 'initial_silence'
]]
df_6 = df_6
df_7 = df_7
```

```python
df = pd.merge(df_1, df_2, how='inner', on='video_id')
df = pd.merge(df, df_3, how='inner', on='video_id')
df = pd.merge(df, df_4, how='inner', on='video_id')
df = pd.merge(df, df_5, how='inner', on='video_id')
df = pd.merge(df, df_6, how='inner', on='video_id')
df = pd.merge(df, df_7, how='inner', on='video_id')

df = df.round({col: 4 for col in df.select_dtypes(include='float').columns})
df
```

```python
df.to_csv('raw_data/final_data.csv', encoding='utf-8-sig', index=False)
```
