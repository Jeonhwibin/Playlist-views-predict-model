```python
import pandas as pd
import emoji
import re
```

```python
# 물음표 또는 느낌표 여부 (있으면 1, 없으면 0)
def has_question_or_exclamation(title):
    title = title.strip()
    return int(('?' in title) or ('!' in title)) # title안에 !나 ?가 있으면 true -> 1로(int)

# 이모지 + 기호 + 카오모지 포함 여부 (있으면 true, 없으면 false)
def has_emoji_or_symbol(title):
    if any(char in emoji.EMOJI_DATA for char in title):
        return True
    if re.search(r'[\u2600-\u26FF\u2700-\u27BF♡♪]', title):
        return True
    if re.search(r'[\(\[][^\)\]]{1,15}[\)\]]', title):
        return True
    return False

# 공백 제외 문자 수
def char_length_no_space(title):
    return len(title.replace(" ", ""))

# 이모지/기호/카오모지 비율 (공백 제외 문자 수 기준)
def emoji_symbol_ratio(title):
    text_no_space = title.replace(" ", "")
    total_len = len(text_no_space)
    emoji_count = sum(char in emoji.EMOJI_DATA for char in text_no_space)
    symbol_count = len(re.findall(r'[\u2600-\u26FF\u2700-\u27BF♡♪]', text_no_space))
    kaomoji_count = len(re.findall(r'[\(\[][^\)\]]{1,15}[\)\]]', text_no_space))
    total_special = emoji_count + symbol_count + kaomoji_count
    return total_special / total_len if total_len > 0 else 0

# 적용
data['has_question_or_exclamation'] = data['title'].apply(has_question_or_exclamation)
data['char_length'] = data['title'].apply(char_length_no_space)
data['has_emoji'] = data['title'].apply(has_emoji_or_symbol)
data['emoji_ratio'] = data['title'].apply(emoji_symbol_ratio)
```

3. openai api 사용을 위해서 한글만 남겨둔 채로 cleaned_data 만들기

```python
import re
import html
from unidecode import unidecode

def safe_unidecode(text):
    result = ""
    for c in text:
        if '\uAC00' <= c <= '\uD7A3':  # 한글
            result += c
        elif c.isdigit() or c.isspace():  # 숫자, 공백 유지
            result += c
        elif c.isascii() or not c.isprintable():
            result += c
        else:
            result += unidecode(c)
    return result

def clean_title_korean_only_strict(title):
    title = html.unescape(title)
    title = safe_unidecode(title)

    # playlist tag 제거
    playlist_patterns = [
        r'\[.*?playlist.*?\]', r'\(.*?playlist.*?\)',
        r'playlist', r'Playlist', r'PLAYLIST'
    ]
    for pattern in playlist_patterns:
        title = re.sub(pattern, '', title, flags=re.IGNORECASE)

    # 영어 제거
    title = re.sub(r'[A-Za-z]', '', title)

    # 이모지 및 모든 특수문자 제거 (한글, 숫자, 공백, ?, !, .제외)
    title = re.sub(r'[^\uAC00-\uD7A30-9\s\?\.\!]', '', title)

    # 공백 정리
    title = re.sub(r'\s+', ' ', title).strip()

    return title
```

```python
data['cleaned_title'] = data['title'].apply(clean_title_korean_only_strict)
```

4. openai api 사용하여 구어체여부, 오감표현 여부 컬럼 만들기

```python
import openai
import pandas as pd
from tqdm import tqdm
from dotenv import load_dotenv
# OpenAI client 설정
load_dotenv()
client = openai.OpenAI(api_key="OPENAI_API_KEY")

# 프롬프트 생성 함수
def make_combined_prompt(text):
    return f"""
당신은 텍스트 언어 분석 전문가입니다.
다음 문장에 대해 두 가지를 판별하세요.

[1] 구어체 판별:
문장의 첫 문장이 일상 대화체나 구어체인지, 아니면 서면체나 문어체(형식적 글쓰기)인지 판단합니다.

판단 기준:
문장이 아래 항목 중 "구어체" 기준을 충족하면 "구어체", 그렇지 않으면 "구어체 아님"으로 판정합니다.
단, 구어체로 판단하려면 기준 중 2개 이상을 충족해야 합니다.
(단, 명령형 또는 감탄사 하나만 있어도 구어체로 판단)

1. "구어체"
- 일상 대화, 사람 간 대화, 말하는 느낌 (반말이면 무조건 구어체)
- 친구, 지인, 팔로워 등을 대상으로 한 친근한 표현
- 구어적 어미 및 반말: ~했어, ~할래, ~봐, ~해 등
- 명령형 표현 포함: 꼭~, 지금~, ~하세요, ~해봐, ~보세요, ~들어봐 등
- 의문형 표현 포함: 어떻게, 왜, 무엇, 뭐, 어디, 누가, 언제, ~인가요, ~일까, ?
- 감탄 표현 포함: 대박, 헐, 와, 우와, 미쳤다, !!, !
- 감탄사 또는 행동 유도 문구: 예) 필수!, 확인하세요!, 들어봐!
- 주어 생략 또는 간결/파편적 문장 (3단어 이하 + 느낌표 또는 물음표 포함)
- 내가, 네가 등 말하는 듯한 표현이 나타날 때

예시)
- 와, 진짜 대박!
- 들어봐, 이건 꼭 알아야 해.
- 지금 시작해봐!
- 왜 그런 걸까?
- 대박이다, 미쳤다!

2. "구어체 아님"
- 위 구어체 기준 중 2개 미만 충족 시
- 정보 전달, 설명형, 나열형 문장
- 제품명, 장소명, 음악 제목, 테마 단어 나열
- 뉴스, 논문, 학술적, 서면적 스타일
- 제목형 문장 (문장 구성 없이 단어만 나열)
- 형식상 "구"나 "절" 형태로 문장이 아닌 경우
  (단, 느낌표 또는 물음표가 있으면 구어체로 판단)

예시)
- 일하면서 듣기 좋은 재즈 플리 중간광고 없음 → 구어체 아님
- 클래식, 재즈, 팝, 락 → 구어체 아님
- 겨울 추천 음악 플레이리스트 → 구어체 아님
- 음악을 들으며 휴식을 → 구어체 아님

추가 규칙:
* 구어체로 판단하는 경우, 모델은 자신의 판단 신뢰도가 0.7 이상일 때만 "구어체"로 최종 판단합니다.
* 판단 신뢰도가 0.7 미만이면 "구어체 아님"으로 판정합니다.


[2] 오감 자극 표현 판별:
문장에서 시각, 촉각, 후각, 미각 중 하나라도 오감을 자극하는 표현이 포함되어 있는지 판단합니다.

판단 기준:
- 반드시 시각, 촉각, 후각, 미각의 4가지 감각 중 하나에 해당해야 합니다.
- 청각(예: 음악, 소리, 듣다, 들린다, 멜로디 등)과 관련된 표현은 "오감 있음"으로 판단하지 않습니다.

(1) 시각 (눈으로 보는 감각)
- 색상, 빛, 형태, 장면을 표현하는 단어
- 예시: 눈부신, 화려한, 선명한, 반짝이는, 붉은, 푸른, 황금빛, 생생한, 투명한, 어두운, 밝은, 푸르른, 다채로운, 빛나는, 반짝임, 색색의

(2) 촉각 (만졌을 때 느끼는 감각)
- 촉감, 온도, 질감을 표현하는 단어
- 예시: 부드러운, 촉촉한, 따뜻한, 시원한, 포근한, 매끄러운, 까슬까슬한, 뻣뻣한, 차가운, 따끈한, 말랑한

(3) 후각 (코로 맡는 향기)
- 향, 냄새, 향기, 냄새의 정도나 종류를 표현
- 예시: 향긋한, 진한 향, 상큼한 향, 고소한 냄새, 구수한 냄새, 새콤한 향, 쿰쿰한, 스모키한 향

(4) 미각 (입으로 느끼는 맛)
- 맛, 풍미, 식감 표현
- 예시: 상큼한, 달콤한, 고소한, 짭짤한, 매콤한, 새콤한, 쌉싸름한, 부드러운 식감, 아삭한, 쫄깃한, 담백한, 고소한 맛, 바삭한

판정 방법:
- 문장에 위 단어군 중 하나라도 포함 → "오감 있음"
- 전혀 포함되지 않음 → "오감 없음"

주의:
- 청각 관련 표현 (예: 듣기, 소리, 음악, 멜로디, 음색 등)은 절대 "오감 있음"으로 판단하지 않습니다.
- 오감과 직접 관련 없는 느낌적 표현 (예: 기분 좋은, 편안한 분위기, 감동적인 등)은 "오감 없음"으로 판단합니다.

예외 예시:
예) 듣기만 해도 기분 좋아지는 재즈와 함께 맞이하는 봄 → 오감 없음 (청각 표현 → 제외)
예) 부드러운 촉감의 니트와 따뜻한 햇살 → 오감 있음 (촉각 + 시각)
예) 향긋한 커피 향과 고소한 빵 냄새 → 오감 있음 (후각 + 미각)
예) 다채로운 색감의 봄꽃 → 오감 있음 (시각)
예) 마음이 편안해지는 풍경 → 오감 없음 (느낌 표현 → 제외)


[응답 형식 예시 (꼭 지키세요)]
구어체: 구어체
오감: 오감 있음

문장:
{text}

당신의 판단:
"""

# 분석 함수
def analyze_text(text):
    try:
        prompt = make_combined_prompt(text)
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            temperature=0
        )
        output = response.choices[0].message.content.strip()
        lines = output.split("\n")
        speech_type, sensory = None, None
        for line in lines:
            if line.lower().startswith("구어체:"):
                speech_type = line.split(":")[1].strip()
            if line.lower().startswith("오감:"):
                sensory = line.split(":")[1].strip()
        return speech_type, sensory
    except Exception as e:
        print(f"❗Error: {e}")
        return None, None

# 중간 저장 + 데이터셋 적용
tqdm.pandas()

save_interval = 500
filename_template = "progressive_save_{:04d}.csv"

results = []
for i in tqdm(range(len(data))):
    text = data.loc[i, 'cleaned_title']
    speech_type, sensory = analyze_text(text)
    results.append((speech_type, sensory))

    # 결과 넣기
    data.loc[i, 'speech_type'] = speech_type
    data.loc[i, 'sensory'] = sensory

    # 500개마다 저장
    if (i + 1) % save_interval == 0 or (i + 1) == len(data):
        file_index = (i + 1) // save_interval
        data.iloc[:i+1].to_csv(filename_template.format(file_index), index=False, encoding='utf-8-sig')
        print(f"✅ Saved up to {i+1} rows → {filename_template.format(file_index)}")

# 최종 확인
print(data.head())
```
