```python
import pandas as pd
import emoji
import re
```

```python
# ë¬¼ìŒí‘œ ë˜ëŠ” ëŠë‚Œí‘œ ì—¬ë¶€ (ìˆìœ¼ë©´ 1, ì—†ìœ¼ë©´ 0)
def has_question_or_exclamation(title):
    title = title.strip()
    return int(('?' in title) or ('!' in title)) # titleì•ˆì— !ë‚˜ ?ê°€ ìˆìœ¼ë©´ true -> 1ë¡œ(int)

# ì´ëª¨ì§€ + ê¸°í˜¸ + ì¹´ì˜¤ëª¨ì§€ í¬í•¨ ì—¬ë¶€ (ìˆìœ¼ë©´ true, ì—†ìœ¼ë©´ false)
def has_emoji_or_symbol(title):
    if any(char in emoji.EMOJI_DATA for char in title):
        return True
    if re.search(r'[\u2600-\u26FF\u2700-\u27BFâ™¡â™ª]', title):
        return True
    if re.search(r'[\(\[][^\)\]]{1,15}[\)\]]', title):
        return True
    return False

# ê³µë°± ì œì™¸ ë¬¸ì ìˆ˜
def char_length_no_space(title):
    return len(title.replace(" ", ""))

# ì´ëª¨ì§€/ê¸°í˜¸/ì¹´ì˜¤ëª¨ì§€ ë¹„ìœ¨ (ê³µë°± ì œì™¸ ë¬¸ì ìˆ˜ ê¸°ì¤€)
def emoji_symbol_ratio(title):
    text_no_space = title.replace(" ", "")
    total_len = len(text_no_space)
    emoji_count = sum(char in emoji.EMOJI_DATA for char in text_no_space)
    symbol_count = len(re.findall(r'[\u2600-\u26FF\u2700-\u27BFâ™¡â™ª]', text_no_space))
    kaomoji_count = len(re.findall(r'[\(\[][^\)\]]{1,15}[\)\]]', text_no_space))
    total_special = emoji_count + symbol_count + kaomoji_count
    return total_special / total_len if total_len > 0 else 0

# ì ìš©
data['has_question_or_exclamation'] = data['title'].apply(has_question_or_exclamation)
data['char_length'] = data['title'].apply(char_length_no_space)
data['has_emoji'] = data['title'].apply(has_emoji_or_symbol)
data['emoji_ratio'] = data['title'].apply(emoji_symbol_ratio)
```

3. openai api ì‚¬ìš©ì„ ìœ„í•´ì„œ í•œê¸€ë§Œ ë‚¨ê²¨ë‘” ì±„ë¡œ cleaned_data ë§Œë“¤ê¸°

```python
import re
import html
from unidecode import unidecode

def safe_unidecode(text):
    result = ""
    for c in text:
        if '\uAC00' <= c <= '\uD7A3':  # í•œê¸€
            result += c
        elif c.isdigit() or c.isspace():  # ìˆ«ì, ê³µë°± ìœ ì§€
            result += c
        elif c.isascii() or not c.isprintable():
            result += c
        else:
            result += unidecode(c)
    return result

def clean_title_korean_only_strict(title):
    title = html.unescape(title)
    title = safe_unidecode(title)

    # playlist tag ì œê±°
    playlist_patterns = [
        r'\[.*?playlist.*?\]', r'\(.*?playlist.*?\)',
        r'playlist', r'Playlist', r'PLAYLIST'
    ]
    for pattern in playlist_patterns:
        title = re.sub(pattern, '', title, flags=re.IGNORECASE)

    # ì˜ì–´ ì œê±°
    title = re.sub(r'[A-Za-z]', '', title)

    # ì´ëª¨ì§€ ë° ëª¨ë“  íŠ¹ìˆ˜ë¬¸ì ì œê±° (í•œê¸€, ìˆ«ì, ê³µë°±, ?, !, .ì œì™¸)
    title = re.sub(r'[^\uAC00-\uD7A30-9\s\?\.\!]', '', title)

    # ê³µë°± ì •ë¦¬
    title = re.sub(r'\s+', ' ', title).strip()

    return title
```
